# **Example Animation**

### Import manim:

```python
from manim import *
```

Example manimation:

```python
%%manim -qm ExampleManimation

class ExampleManimation(Scene):
    def construct(self):
        logo = ImageMobject("media/Vexv5-logo.png").scale(1.2)
        
        title = Text("52800R Manimations", font_size=72)
        subtitle = Text("Explaining Entries Visually", font_size=32, color=GRAY)
        
        banner = ManimBanner().scale(0.2).to_edge(DOWN, buff=0.3)
        
        team_info = VGroup(title, subtitle).arrange(DOWN, buff=0.3)
        
        content = Group(logo, team_info).arrange(DOWN, buff=0.7).move_to(ORIGIN)

        # Animation Sequence
        self.play(
            banner.create(),
            FadeIn(logo, shift=UP * 0.5),
            run_time=1.5
        )
        
        self.play(
            banner.expand(),
            Write(team_info),
            logo.animate.scale(1.1), 
            run_time=2
        )
        

        self.wait(2)

        self.play(
            FadeOut(content, scale=1.2, opacity=0), # "Zoom into camera" exit
            Unwrite(banner),
            run_time=1
        )
```
